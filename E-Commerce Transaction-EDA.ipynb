African E-Commerce Transaction Data Analysis (Synthetic Dataset)

üìù Dataset Description

This dataset simulates real-world e-commerce transactions across selected African countries. It contains 10,000 transaction-level records designed to reflect realistic purchasing behavior, customer demographics, operational variables, and common data quality issues.

The dataset represents online retail activity between 2022 and 2023, capturing customer purchases across multiple product categories and payment methods.

Each row represents one transaction, including product information, customer details, pricing, and shipping attributes.

The dataset intentionally includes controlled inconsistencies and data quality issues to simulate real-world analytical challenges.

üåç Market Context

The dataset focuses on African markets including:

Nigeria

Kenya

Ethiopia

South Africa

Ghana

Egypt

Some country names are intentionally inconsistent (e.g., "NIGERIA", "Kenya ", "S. Africa") to simulate real-world data entry issues.

üì¶ Dataset Structure

The dataset contains the following columns:

| Column Name      | Description                                            |
| ---------------- | ------------------------------------------------------ |
| transaction_id   | Unique identifier for each transaction                 |
| customer_id      | Unique identifier for customers                        |
| order_date       | Date of transaction (mixed formats)                    |
| product_category | Product category (Electronics, Clothing, Home, Sports) |
| product_price    | Price per unit                                         |
| quantity         | Number of units purchased                              |
| payment_method   | Payment type used                                      |
| country          | Customer country                                       |
| discount_percent | Discount applied to transaction                        |
| shipping_days    | Delivery time in days                                  |
| customer_age     | Age of customer                                        |
| customer_gender  | Gender of customer                                     |
| is_returned      | Whether the item was returned (0/1)                    |

What drives revenue across countries and categories?

How do customer demographics influence purchasing behavior?

What is the impact of discounts on sales?

Are there seasonal trends in transactions?

What patterns exist in product returns?

Which markets show stronger growth or operational efficiency?
